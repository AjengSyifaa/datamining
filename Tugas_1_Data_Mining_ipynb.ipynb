1. Memuat data
- Import library yang diperlukan, seperti Pandas.
- Muat dataset dari file CSV ke dalam DataFrame.

```python
import pandas as pd
df = pd.read_csv('/content/drive/MyDrive/Semester 5/Data Mining/Tugas Praktek 1/movie_sample_dataset.csv')
df.head()
```

2. Memeriksa data
- Tampilkan beberapa baris pertama dari dataset untuk memahami struktur data


```python
#Tampilkan beberapa baris pertama dari dataset untuk memahami struktur data
df.head(1)
```

- Periksa informasi umum tentang dataset, termasuk tipe data dan nilai yang hilang.

```python
df.dtypes
```

```python
import numpy as np

# Mengganti " " dengan NaN
df.replace(" ", np.nan, inplace=True)

# Memeriksa jumlah missing values di setiap kolom
print(df.isnull().sum())

```

3. Membersihkan data

```python
#Hapus baris yang memiliki nilai NaN di kolom penting seperti gross dan budget.

# Menghapus baris dengan missing values pada kolom
df.dropna(subset=["gross", "budget"], axis=0, inplace=True)

# Reset index setelah menghapus baris
df.reset_index(drop=True, inplace=True)

print(df.head())
```

```python
#Atasi nilai yang tidak konsisten atau kesalahan penulisan di kolom, seperti perbedaan antara "Color" dan "color".
df['country'] = df['country'].str.upper()
df['color'] = df['color'].str.capitalize()

df.head()
```

```python
# Ubah atau hapus nilai-nilai yang tidak standar seperti nilai negatif atau "N/A"
import numpy as np

# Mengganti "-, N/A" dengan NaN
df.replace("- , N/A", np.nan, inplace=True)

print(df.head())
```

4. Transforming data

Ubah tipe data kolom menjadi tipe data yang sesuai (misalnya, konversi budget dan gross menjadi numerik).

```python
# Ubah tipe data kolom menjadi tipe data yang sesuai

df[["duration", "gross", "title_year", "budget",
"movie_facebook_likes"]] = df[["duration", "gross", "title_year",
 "budget", "movie_facebook_likes"]].astype(int)

df[["imdb_score"]] = df[["imdb_score"]].astype(float)
df.dtypes
```

Pisahkan genre yang bergabung dalam satu kolom menjadi beberapa kolom terpisah
(opsional).

```python
df['genres'].str.split()

df['genres'].str.split(pat = '|', expand = True)
```

5. Penyimpanan data

```python
df.to_csv('movie_dataset_cleaned.csv', index=False)
```
